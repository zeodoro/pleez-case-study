```
import streamlit as st
import pandas as pd

data = {
    "item": ["Burger", "Sushi", "Pizza"],
    "price": [10.0, 14.0, 12.0],
    "cost": [4.5, 5.0, 4.0],
    "sales": [180, 120, 200],
}

df = pd.DataFrame(data)
df["margin"] = (df["price"] - df["cost"]) * df["sales"]

st.title("Menu Optimization Demo")
st.dataframe(df)
```
